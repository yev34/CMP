```python
import math
import numpy as np 
import matplotlib.pyplot as plt 
def f(x,y):
    return x+math.cos(y/math.sqrt(3))
x0=1.2
b=2.2
h=0.1

x=np.arange(x0,b+h,h)
n=len(x)-1
y=np.empty(n+1)
y[0]=2.1

for i in range(n):
    y[i+1]=y[i]+f(x[i],y[i])*h
y_rounded=np.round_(y,4)
print ('x=',x,'\ny=',y_rounded)
plt.plot(x,y,'o',x,y,'red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Метод Ейлера')
plt.legend(['точки','x+cos(y/sqrt(3))'])
plt.grid()
plt.show()
```

```python
def f(x,y):
    return x+math.sin(y/math.sqrt(2))
x0=0.8
b=1.8
h=0.1

x=np.arange(x0,b+h,h)
n=len(x)-1
y=np.empty(n+1)
y[0]=1.3

for i in range(n):
     y[i+1]=y[i]+(f(x[i],y[i])+f(x[i+1],y[i]+h*f(x[i],y[i])))*h/2 
y_rounded=np.round_(y,4)
print ('x=',x,'\ny=',y_rounded)
plt.plot(x,y,'o',x,y,'red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Метод Ейлера')
plt.legend(['точки','x+sin(y/sqrt(2))'])
plt.grid()
plt.show()
```
